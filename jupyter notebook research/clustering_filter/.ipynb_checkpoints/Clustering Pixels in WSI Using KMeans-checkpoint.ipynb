```python
import os
import cv2
import tensorflow as tf
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
```

```python
input_img = '/code/research/images/mcw_he_stylegan_fakes/random_9_480.png'
input_img = '/code/research/images/mcw_he_stylegan_fakes/random_7_980.png'
```

```python
img = cv2.imread(input_img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img_pixel_list = img.reshape((img.shape[0] * img.shape[1], 3))
```

```python
img_pixel_list.shape
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(2)
```

```python
type(cluster)
```

```python
cluster.fit(img_pixel_list)
```

```python
pixel_labels = cluster.labels_
```

```python
img_mask = pixel_labels.reshape((img.shape[0],img.shape[1]))
```

```python
img_mask.shape
```

```python
plt.imshow(img_mask)
```
